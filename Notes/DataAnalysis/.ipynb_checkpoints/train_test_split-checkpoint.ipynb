```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
df = pd.read_csv("wrangled_auto.csv")
```

```python
Y=df['price']
X=df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0) #random_state is a random seed.
```

Cross Validation is to break up the data into equal "folds" then use different folds as training and testing data, to be able to compare the error rates of each with each other.

```python
lr = LinearRegression()
```

```python
#r2 scores for each fold
scores = cross_val_score(lr, X, Y, cv=3) #Cross Validate 3 folds
print(scores)
```

```python
#out of sample r2
np.mean(scores)
```

```python
yhat=cross_val_predict(lr,X,Y,cv=3)
```

```python
print(yhat)
```
