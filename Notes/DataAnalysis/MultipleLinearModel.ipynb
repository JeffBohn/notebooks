```python
import pandas as pd
import numpy as np
#import seaborn as sns
%matplotlib inline
#import matplotlib
from matplotlib import pyplot as plt
#import scipy.stats as stats
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("wrangled_auto.csv")
```

```python
lm=LinearRegression()
```

```python
Y=df['price']
Z=df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
```

```python
lm.fit(Z, Y)
```

```python
Yhat=lm.predict(Z)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
plt.title("Price predicted based on Highway MPG")
#plt.scatter(range(Z.shape[0]), Y, label="actual")
#plt.scatter(range(Z.shape[0]), Yhat, label="predicted")
plt.plot(Y, label="actual")
plt.plot(Yhat, label="predicted")
plt.legend()
plt.xlabel('Sample Number')
plt.ylabel('Price')
plt.grid()
plt.show()
```
