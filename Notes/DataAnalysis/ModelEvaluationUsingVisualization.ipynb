```python
import pandas as pd
#import numpy as np
import seaborn as sns
%matplotlib inline
#import matplotlib
from matplotlib import pyplot as plt
#import scipy.stats as stats
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("wrangled_auto.csv")
```

<h2>Regression Plot</h2>

```python
sns.regplot(x="highway-mpg",y="price",data=df)
plt.ylim(0,)
plt.show()
```

<h2>Residual Plots</h2>
Displays the differences between the predicted and actual values.

```python
#First generate a simple linear regression model to evaluate with the residual plot
lm=LinearRegression()
X=df[['highway-mpg']]
Y=df['price']
lm.fit(X,Y)
Yhat=lm.predict(X)
```

```python
#Calculate the error values of the model.
Error = Y - Yhat
#Error = (Error-Error.mean())/Error.std() #z-score
```

<h3>Attempt at do-it-myself</h3>

```python
plt.title = "Residual Plot"
plt.scatter(X, Error, label="error")
plt.legend()
plt.xlabel('Highway MPG')
plt.ylabel('Price Error')
plt.show()
```

<h3>Using Seaborn</h3>

```python
sns.residplot(df['highway-mpg'], df['price'])
plt.show()
```

<h3>The way lesson DA0101EN did it</h3>

```python
width=12
height=10
plt.figure(figsize=(width,height))
sns.residplot(df['highway-mpg'], df['price'])
plt.show()
```

<h3>Analysis</h3>
The above residual plot shows a visible curvature.  Because this is non-linear, it suggest our simple linear model may not be very good at prediction.

<h2>Distribution Plots</h2>
<ul>
<li>Counts the predicted value versus the actual value, basically breaking the results of each into equal sized "bins" then comparing them.
<li>Useful for visualizing models with more than one independent value.
</ul>

<h3>Show it with the "bins" or histograph.</h3>

```python
ax1 = sns.distplot(df['price'], hist=True, color="r", label="Actual Value")
sns.distplot(Yhat, hist=False, color="b", label="Fitted Values", ax = ax1)
plt.show()
```

<h3>Continuous versus discrete values</h3>
In this case the Price is continuous so a histograph is not best.  Pandas was already converting to a distribution represented by that red line above.  By scaling the Y axis to make the area under the red line equal to one.  But we can avoid the histograph being displayed by setting hist=false.

```python
ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(Yhat, hist=False, color="b", label="Fitted Values", ax = ax1)
plt.show()
```

<h3>Analysis</h3>
In this example, the model does not predict the values over 30K very well at all, and the others are offset a bit.  This could mean using a single linear model is not adequate.

<h3>Try again with a multiple linear model</h3>

```python
Ym=df['price']
Zm=df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
lm.fit(Zm, Ym)
Yhatm=lm.predict(Zm)

ax1m = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(Yhatm, hist=False, color="b", label="Fitted Values", ax = ax1m)
plt.show()
```
