```python
import pandas as pd
import numpy as np
#import seaborn as sns
%matplotlib inline
#import matplotlib
from matplotlib import pyplot as plt
#import scipy.stats as stats
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("wrangled_auto.csv")
```

```python
lm=LinearRegression()
```

```python
X=df[['highway-mpg']]
Y=df['price']
```

```python
lm.fit(X,Y)
```

```python
Yhat=lm.predict(X)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
plt.title("Price predicted based on Highway MPG")
plt.scatter(X, Y, label="actual")
plt.scatter(X, Yhat, label="predicted")
plt.legend()
plt.xlabel('Highway MPG')
plt.ylabel('Price')
plt.grid()
plt.show()
```
