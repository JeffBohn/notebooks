```python
import pandas as pd
import numpy as np
#import seaborn as sns
%matplotlib inline
#import matplotlib
from matplotlib import pyplot as plt
#import scipy.stats as stats
#from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("wrangled_auto.csv")
```

<h2>function for plotting</h2>

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ ' + Name)
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

<h2>Polynomial Functions in various degrees</h2>

```python
x=df['highway-mpg']
y=df['price']
```

```python
#3rd order
f=np.polyfit(x,y,3)
p=np.poly1d(f)
print(p)
```

```python
PlotPolly(p,x,y,'highway-mpg 3')
```

```python
#11th order
f=np.polyfit(x,y,11)
p=np.poly1d(f)
print(p)
```

```python
PlotPolly(p,x,y,'highway-mpg 11')
```

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y, p(x))
print('The R-square value is: ', r_squared)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(df['price'], p(x))
```
